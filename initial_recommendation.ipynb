```python
import pandas as pd
import numpy as np

anime_cols = ['anime_id', 'name', 'genre', 'type', 'episodes', 'rating']
anime = pd.read_csv('/Users/brangmai/Desktop/Capstone/modified_data/anime.csv', names=anime_cols, usecols=range(6), encoding="ISO-8859-1")
anime = anime[1:]
anime = pd.DataFrame(anime)
anime.head(-20)
#print(len(anime))
```

# The highest rated animes from all type

```python
highest_rated_animes = anime.sort_values(by='rating', ascending=False)
highest_rated_animes = highest_rated_animes.head(10)

highest_rated_animes.reset_index(drop=True, inplace=True)
cols = list(highest_rated_animes.columns.values)

highest_rated_animes = highest_rated_animes[cols[1:4] + [cols[-1]]]

#highiest_rated_animes.style.set_properties(subset=['name'], **{'text-align': 'left'})
highest_rated_animes.head(50)

# for index, row in highiest_rated_animes.iterrows():
#     print((index + 1), row['name'], row['genre'])
```

# The highest rated movies

```python
anime_movies = anime.loc[anime['type'] == 'Movie']
anime_movies = anime_movies.sort_values(by='rating', ascending=False)
anime_movies.reset_index(drop=True, inplace=True)
cols = list(anime_movies.columns.values)
anime_movies = anime_movies[cols[1:3] + [cols[-1]]]
anime_movies.head(10)
```

# The highest rated TV-series

```python
anime_tv = anime.loc[anime['type'] == 'TV']
anime_tv = anime_tv.sort_values(by='rating', ascending=False)
anime_tv.reset_index(drop=True, inplace=True)

cols = list(anime_tv.columns.values)
anime_tv = anime_tv[cols[1:3] + [cols[-1]]]
anime_tv.head(10)
```

# The highest rated original video animation (neither movies nor tv )

```python
anime_others = anime.loc[(anime['type'] != 'Movie') & (anime['type'] != 'TV')]
anime_others = anime_others.sort_values(by='rating', ascending=False)
anime_others.reset_index(drop=True, inplace=True)

cols = list(anime_others.columns.values)
anime_others = anime_others[cols[1:3] + [cols[-1]]]
anime_others.head(10)
```
