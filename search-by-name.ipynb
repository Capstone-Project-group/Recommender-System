```python
import pandas as pd
import numpy as np


rating_cols = ['user_id', 'anime_id', 'rating']
ratings = pd.read_csv('/Users/brangmai/Desktop/Capstone/modified_data/rating.csv', names=rating_cols, usecols=range(3), encoding="ISO-8859-1")
ratings = ratings[1:]
#ratings.head(50)

anime_cols = ['anime_id', 'name', 'genre', 'type', 'episodes', 'rating']
anime = pd.read_csv('/Users/brangmai/Desktop/Capstone/modified_data/anime.csv', names=anime_cols, usecols=range(6), encoding="ISO-8859-1")
anime = anime[1:]

rated_anime = pd.merge(anime, ratings)
rated_anime.head(10)
```

```python
anime_type = []
for type_item in rated_anime['type']:
    if type_item not in anime_type:
        anime_type.append(type_item)
        
        
for item in anime_type:
    print(item)
```

```python
# SEARCH BY TITLE
search_by_title = input("Please, enter anime title: ")

search_title_exist = False
for ani_name in rated_anime['name']:
    if ani_name == search_by_title:
        search_title_exist = True
        print(f'{search_by_title} is found.')
        break

if not search_title_exist:
    print(f'{search_by_title} cannot be found.')
    
```

```python
anime_ratings = rated_anime.pivot_table(index='user_id', columns=['name'], values='rating')
anime_ratings.head(10)
```

```python
sword_art_ratings = anime_ratings['Sword Art Online: Extra Edition']
sword_art_ratings = sword_art_ratings.dropna()
print('size is: ', sword_art_ratings.size)
sword_art_ratings.head(10)
```

```python
similar_anime = anime_ratings.corrwith(sword_art_ratings)
#similar_anime = similar_anime.dropna()
data_frame = pd.DataFrame(similar_anime)
data_frame.head(10)
```

```python
similar_anime.sort_values(ascending=False)
```

```python
import numpy as np

anime_stats = rated_anime.groupby('name').agg({'rating': [np.size, np.mean]})
anime_stats.head(10)
```

```python
popular_anime = anime_stats['rating']['size'] >= 5
popular_anime = anime_stats[popular_anime].sort_values([('rating', 'mean')], ascending=False)
popular_anime.head(10)
```

```python
data_frame = anime_stats[popular_anime].join(pd.DataFrame(similar_anime, columns=['similarity']))
data_frame.head(10)
```
